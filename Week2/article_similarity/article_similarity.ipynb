# Article Similarity

This notebook implements the following:
- Read CSV using csv module
- Clean text
- Build global vocabulary
- Binary vectors
- Cosine similarity using numpy
- Save pickle file
- Get top 3 similar articles

```python
import csv
import re
import numpy as np
import pickle
```

## 1. Read CSV file

```python
articles = []

with open('articles.csv', encoding='utf-8') as f:
    for r in csv.DictReader(f):
        articles.append(r)

articles
```

## 2. Clean text and build vocabulary

```python
def clean(t):
    t = t.lower()
    t = re.sub(r'[^\w\s]', '', t)
    t = re.sub(r'\d+', '', t)
    return t.split()

vocab = sorted(set(w for a in articles for w in clean(a['content'])))
vocab
```

## 3. Build binary vectors

```python
X = np.array([[1 if w in clean(a['content']) else 0 for w in vocab] for a in articles])
X
```

## 4. Cosine similarity

```python
dot = np.dot(X, X.T)
norm = np.outer(np.linalg.norm(X, axis=1), np.linalg.norm(X, axis=1))
sim = dot / norm

sim
```

## 5. Save similarity matrix

```python
pickle.dump(sim, open('similarities.pkl', 'wb'))
```

## 6. Top 3 similar articles

```python
def top3(article_id):
    idx = next(i for i, a in enumerate(articles) if a['id'] == str(article_id))
    order = np.argsort(sim[idx])[::-1]
    return [articles[i]['title'] for i in order if i != idx][:3]

top3(1)
```
